```python
import pandas as pd
import numpy as np
import re
import json

movies=pd.read_csv('/home/silvia/corsi/Informatica/Python/data_science_foundations/movies/tmdb_5000_movies.csv')
movies
```

```python
credits=pd.read_csv('/home/silvia/corsi/Informatica/Python/data_science_foundations/movies/tmdb_5000_credits.csv')
credits
```

```python
# For each movie, compute the number of cast members

def extract_cast(x):
    d=json.loads(x)
    if len(d)==0 :
        nc=None
    else :
        nc=pd.DataFrame(d)['cast_id'].count()
    return nc

movies.rename(columns={'id': 'movie_id'}, inplace=True) # inplace rinomina la colonna del database senza farne una copia
movies

cast=movies.merge(credits,on=['movie_id'],suffixes=['',' credits'])[['movie_id','original_title','cast']]
cast

#forse c'Ã¨ qualche cast nullo e non funzia l'applicazione della funzione
cast['n. cast members']=credits['cast'].apply(lambda row: extract_cast(row) if(pd.notnull(row)) else None)
cast
```
